```python
import pandas as pand
import numpy as num
import seaborn as sns
import matplotlib.pyplot as plt
data = pand.read_csv('anime1.csv')
data
#Вывод датасета
```

```python
#Данные о столбиках(название, типы данных и т.д.)
data.info()
#name - Название
#genre - Жанр
#type - Тип
#episodes - Эпизоды
#rating - Рейтинг
#members - Количество зрителей
```

```python
#Столбчатая диаграмма, показывающая количество разных жанров аниме
data['type'].hist(figsize=(16, 8))	
plt.title('Количество разных жанров аниме')
plt.show()
#Гипотеза: Фильмов больше, чем Ova(Опровержено)
```

```python
#График, показывающий рейтинг аниме
plt.figure(figsize=(10,4), dpi = 150)
plt.plot(data['rating'].head(500))
plt.xlabel('Счет')
plt.ylabel('Рейтинг аниме')
plt.title('Рейтинг аниме')
plt.grid()
plt.show()
```

```python
#Данные о столбике рейтинга
data['rating'].describe().loc[['count', 'mean', 'min', 'max']].T
```

```python
#Корреляция рейтинга и зрителей
data.corr()['rating'].sort_values()[1:-1]
```

```python
#Визуализация корреляции
plt.figure(figsize=(10,6),dpi=200)
sns.heatmap(data.corr()[1:], annot=True)
```

```python
#График зависимости рейтинга и зрителей
plt.figure(figsize=(10,4))
sns.relplot(x = data['members'], y = data['rating'])
#plt.plot(data['members'].head(500), data['rating'].head(500))
plt.xlabel('Количество зрителей')
plt.ylabel('Рейтинг аниме')
plt.grid()
plt.show()
```
